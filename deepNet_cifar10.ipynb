```lua
require "nn"
require "cunn"
Plot = require "itorch.Plot"
utils = require "cifar10-utils"
```

```lua
data = require("load_cifar-10")
trainset = data.trainset
testset = data.testset
classes = data.classes
```

```lua
utils.net = require "deepNet"

utils.criterion = nn.ClassNLLCriterion()

if utils.opt.cuda == true then
utils.net:cuda()
utils.criterion:cuda()
end

utils.parameters, utils.gradParameters = utils.net:getParameters()
```

```lua
utils.net = torch.load("deep_net")
utils.criterion = nn.ClassNLLCriterion()

if utils.opt.cuda == true then
utils.net:cuda()
utils.criterion:cuda()
end

utils.parameters, utils.gradParameters = utils.net:getParameters()
```

```lua
train_losses = {0}
test_losses = {0}
```

```lua
utils.opt.optimState.learningRate = 1e-4
utils.opt.optimState.weightDecay = 1e-10
utils.opt.optimState.momentum = 0.9
```

```lua
iterations = 8000
utils.err.any = false
if #train_losses == 0 then
    train_losses[1] = -math.log(0.1)
    test_losses[1] = -math.log(0.1)
end
local loss_plot = Plot():line(torch.range(1,#train_losses),train_losses):title("Optimizing... batch 0")
loss_plot:line(torch.range(1,#test_losses),test_losses, "blue")
loss_plot:draw()
for i = 2, iterations do
    local _new_params, minibatch_loss = utils.update()
    if utils.err.any == true then
        print("stoping because there was an error")
        break
    end
    if i % 2000 == 0 then
      train_losses[#train_losses + 1] = utils.evaluate(trainset) -- append the new loss      
      test_losses[#test_losses +1], pred_hist = utils.evaluate(testset)
      print(string.format("minibatches processed: %6s, loss = %6.6f, test_loss = %6.6f", i, train_losses[#train_losses], test_losses[#test_losses]))
      print(pred_hist:resize(1,10))
      loss_plot:title("Optimizing... batch "..i)
      loss_plot:line(torch.range(1,#train_losses),train_losses, "red")
      loss_plot:line(torch.range(1,#test_losses),test_losses, "blue"):redraw()
      os.execute("mpg123 ~/workspace/SVHN_torch/sounds/ping.mp3 &")
    end
end
loss_plot:title("Optimitzation done"):redraw()
os.execute("mpg123 ~/workspace/SVHN_torch/sounds/ping.mp3")
os.execute("mpg123 ~/workspace/SVHN_torch/sounds/ping.mp3")
--os.execute("mpg123 ~/workspace/SVHN_torch/sounds/message.mp3")
collectgarbage()
```
